```python
# .transform
```

```python
dftranform = sns.load_dataset('tips')
dftranform
```

```python
dftranform['medianbill'] = df.groupby('time')['total_bill'].transform('median')
dftranform
```

```python
dftranform.query('time == "Lunch"')
```

```python
dftranform.query('total_bill > medianbill')
```

```python
dftranform[dftranform['total_bill'] > dftranform['medianbill']]
```

```python
#lambda zscore
```

```python
dftranform['zscore'] = df.groupby('time')['total_bill'].transform(lambda x: (x-x.mean())/ x.std())
```

```python
dftranform
```

```python
#apply
```

```python
dftranform.groupby('time').apply(lambda x: (x['tip'] / x['total_bill']).mean()).to_frame('avgtippercent').reset_index()
```

```python
# % tip 
```

```python
dftranform.groupby('time').apply(lambda x: 100 * (x['tip'] / x['total_bill']).mean()).to_frame('avgtippercent').reset_index()
```

```python
# output the avarage total for each taxi color to each row
```

```python
tx['avgtotal'] = tx.groupby('color')['total'].transform('mean')
```

```python
tx
```

```python
#output the row that have a total greater than the avg color
```

```python
tx.query('total > avgtotal')
```

```python
# mean, max, fare for payment type and taxi color
```

```python
tx.groupby(['payment', 'color'])['fare'].agg(avg = 'mean', minimum = 'min', maximum = 'max')
```
